```python
import yfinance as yf
btc = yf.download("BTC-USD", start="2020-01-01", end="2025-01-01")["Close"]
mstr = yf.download("MSTR", start="2020-01-01", end="2025-01-01")["Close"]
```

```python
import pandas as pd
btc_df = pd.DataFrame(btc)
btc_df.tail(10)
```

```python
mstr_df = pd.DataFrame(mstr)
mstr_df.tail(10)
```

```python
df = pd.concat([btc_df, mstr_df], axis=1)
df.columns = ["BTC", "MSTR"]
df = df.dropna()
df.tail(10)
```

```python
import matplotlib.pyplot as plt
df['MSTR_returns'] = (1 + df['MSTR'].pct_change()).cumprod() - 1
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['MSTR_returns'], label="Індекс зростання акцій MSTR")
plt.title("Графік індексу зростання акцій MicroStrategy")
plt.ylabel("Зміна від початкового рівня")
plt.xlabel("Рік")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = df['BTC'].values.reshape(-1, 1)
y = df['MSTR'].values
model = LinearRegression()
model.fit(X, y)
beta = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)
correlation = df['BTC'].corr(df['MSTR'])
print(f"Коефіцієнт нахилу (β): {beta:.4f}")
print(f"Вільний член (α): {intercept:.4f}")
print(f"Коефіцієнт детермінації (R²): {r_squared:.4f}")
print(f"Коефіцієнт кореляції: {correlation:.4f}")
```

```python
import seaborn as sns
sns.regplot(x='BTC', y='MSTR', data=df, scatter_kws={'alpha':0.5})
plt.title("Регресійна залежність ціни MSTR від курсу BTC")
plt.xlabel("Курс BTC")
plt.ylabel("Ціна MSTR")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
va_part = 23909373 / 25843685
print(f"Частка віртуальних активів в сумі активів MicroStrategy складає : {va_part}")
```

```python
btc_share_monobank_scenarios = [5, 10, 15, 20, 25] 
r_squared_monobank = [r_squared * share / va_part for share in btc_share_monobank_scenarios]
correlation_monobank = [correlation * (share / va_part)**0.5 for share in btc_share_monobank_scenarios]

plt.figure(figsize=(10, 5))
plt.plot(btc_share_monobank_scenarios, r_squared_monobank, marker='o', label="Скоригований R² для Monobank")
plt.plot(btc_share_monobank_scenarios, correlation_monobank, marker='s', label="Скоригований коефіцієнт кореляції для Monobank")
plt.xlabel("Частка криптоактивів у Monobank (%)")
plt.ylabel("Значення")
plt.title("Сценарне моделювання впливу BTC на фінансові показники Monobank")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
